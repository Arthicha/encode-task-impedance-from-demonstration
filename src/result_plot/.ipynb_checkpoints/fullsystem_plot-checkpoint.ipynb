```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = '../../resources/demo_data/weight_1.npy'
data = np.load(filepath)
data = data[10:].transpose()
data.shape
```

##### define index

```python
x = 0
y = 1
z = 2
x_ref = 3
y_ref = 4
z_ref = 5
fz = 11
kpz = 20
kvz = 29
imz = 38
ts = np.linspace(0,data[z].shape[0]*0.01,data[z].shape[0])
```

# Z position

```python
plt.rcParams["figure.figsize"] = (10,3)
plt.rcParams['figure.dpi'] = 140
plt.plot(ts,data[z],label="actual z")
plt.plot(ts,data[z_ref],ls='--',label="desired z")
plt.legend()
plt.xlabel("time (sec)",fontsize="x-large")
plt.ylabel("z (meter)",fontsize="x-large")
plt.show()
```

# Z force

```python
plt.rcParams["figure.figsize"] = (10,3)
plt.rcParams['figure.dpi'] = 140
plt.plot(ts,data[fz],label="force z")
plt.legend()
plt.xlabel("time (sec)",fontsize="x-large")
plt.ylabel("force (N)",fontsize="x-large")
plt.show()
```

# Z parameters

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['figure.dpi'] = 140
plt.subplot(3,1,1)
plt.plot(ts,data[kpz])
plt.xlabel("time (sec)",fontsize="x-large")
plt.ylabel("stiffness",fontsize="x-large")
plt.subplot(3,1,2)
plt.plot(ts,data[kvz])
plt.xlabel("time (sec)",fontsize="x-large")
plt.ylabel("damping",fontsize="x-large")
plt.subplot(3,1,3)
plt.plot(ts,data[imz])
plt.xlabel("time (sec)",fontsize="x-large")
plt.ylabel("inertia",fontsize="x-large")
plt.show()
```

# Trajectory & parametes

```python
fig = plt.figure()
skip = 20
plt.rcParams["figure.figsize"] = (10,10)
ax = plt.axes(projection='3d')
ax.plot3D(data[x],data[y],data[z],label="actual")
ax.plot3D(data[x_ref],data[y_ref],data[z_ref],label="desired",ls='--')
ax.scatter(data[x][::skip],data[y][::skip],data[z][::skip], s=0.001*np.power(data[kpz][::skip],2),alpha=0.3)
ax.axes.set_xlim3d(left=-0.25+0.02, right=-0.15+0.02) 
ax.axes.set_ylim3d(bottom=-0.25+0.02, top=-0.15+0.02) 
ax.axes.set_zlim3d(bottom=0.4, top=0.8) 
```
