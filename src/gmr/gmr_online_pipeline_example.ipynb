This is an example of how to use the online GMR.

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from gaussian_mixture_regression import load_GMM, predict_GMR
```

Load in the already trained model.
Currently, it has support only for time or force, not both at the same time :(.   --it should be done by tuesday the full support--

```python
timed_GMM = load_GMM(filepath=str(Path().resolve()), filename='timed_GMM')
```

In a time loop, obtain all the prediction as each time step.
Plot all the predictions for each of dimension.

```python
time = np.linspace(0, 10, 300)
for j in range(21):
    for i in time:
        ni = i/np.amax(time)
        out = predict_GMR(gmm=timed_GMM, timestamp=ni)
        plt.scatter(i, out[j])
    plt.show()
print(out.shape)
```
